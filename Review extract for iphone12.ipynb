```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_soup(url):
    page=requests.get(url).text
    soup=bs(page,'html.parser')
    return soup
review_list=[]
```

```python
def get_reviews(soup):
        reviews= soup.find_all("div",{"data-hook":"review"})
        try:
            for item in reviews:
                review = {
                "r_name" : item.find("span", {"class":"a-profile-name"}).text.strip(),
                "r_rating" : item.find("i", {"data-hook" : "review-star-rating"}).text,
                "r_review ": item.find("span", {"data-hook":"review-body"}).text
                 }
                review_list.append(review)
        except:
            pass
```

```python
review=[]
```

```python
for i in range(1,150):
    soup=get_soup(('https://www.amazon.in/New-Apple-iPhone-12-64GB/product-reviews/B08L5TGWD1/ref=cm_cr_getr_d_paging_btm_next_6?ie=UTF8&reviewerType=all_reviews&pageNumber=')+str(i))
    get_reviews(soup)
    review=review+review_list
    review_list=[]
    print(len(review))
```
